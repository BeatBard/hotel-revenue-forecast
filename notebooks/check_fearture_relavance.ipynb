```python
impact_flags = [
    "IsRamadan", "IsEid", "IsPreRamadan", "IsPostRamadan", "IsLast10Ramadan",
    "IsDSF", "IsSummerEvent", "IsWorldCup", "IsNationalDay", "IsNewYear",
    "IsMarathon", "IsGITEX", "IsFilmFestival", "IsAirshow", "IsArtDubai",
    "IsFoodFestival", "IsPreEvent", "IsPostEvent", "TourismIntensity",
    "RevenueImpact", "IslamicPeriod"
]

def uplift_table(flag):
    base = df['CheckTotal'].mean()
    out = (df.groupby(flag)['CheckTotal']
             .agg(['count', 'mean', 'median', 'std'])
             .rename(columns={'count': 'n'}))
    out['uplift_vs_base_%'] = (out['mean'] - base) / base * 100
    out.index.name = flag
    return out.sort_values("uplift_vs_base_%", ascending=False)
```

```python
all_uplifts = []

for flag in impact_flags:
    try:
        table = uplift_table(flag).reset_index()
        table['Feature'] = flag
        all_uplifts.append(table)
    except Exception as e:
        print(f"⚠️ Skipped {flag}: {e}")
```

```python
# Combine all
uplift_summary = pd.concat(all_uplifts, ignore_index=True)
```

```python
# Save as flat file or view
uplift_summary_sorted = uplift_summary.sort_values(by="uplift_vs_base_%", ascending=False)
print("\n📊 Consolidated Uplift Summary Across All Event Features:")
print(uplift_summary_sorted[["Feature", uplift_summary_sorted.columns[0], "mean", "uplift_vs_base_%"]].head(25))
```

```python
# 🧠 LLM-Ready Revenue Impact Summary for Categorical Features (All Centers)

import os

# ------------------------------------------------------------
# 1. SETUP
# ------------------------------------------------------------
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Directory where revenue center CSVs are stored
input_dir = "revenue_center_data"
output_dir = "uplift_outputs"
os.makedirs(output_dir, exist_ok=True)

impact_flags = [
    "IsRamadan", "IsEid", "IsPreRamadan", "IsPostRamadan", "IsLast10Ramadan",
    "IsDSF", "IsSummerEvent", "IsWorldCup", "IsNationalDay", "IsNewYear",
    "IsMarathon", "IsGITEX", "IsFilmFestival", "IsAirshow", "IsArtDubai",
    "IsFoodFestival", "IsPreEvent", "IsPostEvent", "TourismIntensity",
    "RevenueImpact", "IslamicPeriod"
]

def uplift_table(df, flag):
    base = df['CheckTotal'].mean()
    out = (df.groupby(flag)['CheckTotal']
             .agg(['count', 'mean', 'median', 'std'])
             .rename(columns={'count': 'n'}))
    out['uplift_vs_base_%'] = (out['mean'] - base) / base * 100
    out.index.name = flag
    return out.sort_values("uplift_vs_base_%", ascending=False)

# ------------------------------------------------------------
# 2. LOOP THROUGH REVENUE CENTER FILES
# ------------------------------------------------------------
summary_all_centers = []

for file in os.listdir(input_dir):
    if file.endswith(".csv"):
        center_name = file.replace("_data.csv", "")
        df = pd.read_csv(os.path.join(input_dir, file), parse_dates=['Date'])

        all_uplifts = []
        for flag in impact_flags:
            if flag in df.columns:
                try:
                    table = uplift_table(df, flag).reset_index()
                    table['Feature'] = flag
                    table['RevenueCenter'] = center_name
                    all_uplifts.append(table)
                except Exception as e:
                    print(f"⚠️ Skipped {flag} in {center_name}: {e}")

        if all_uplifts:
            center_uplift_df = pd.concat(all_uplifts, ignore_index=True)
            summary_all_centers.append(center_uplift_df)

            # Save per-center CSV
            center_uplift_df.to_csv(os.path.join(output_dir, f"uplift_{center_name}.csv"), index=False)
            print(f"✅ Saved uplift summary for {center_name}")

# ------------------------------------------------------------
# 3. GLOBAL SUMMARY ACROSS ALL CENTERS
# ------------------------------------------------------------
if summary_all_centers:
    final_summary = pd.concat(summary_all_centers, ignore_index=True)
    final_summary = final_summary.sort_values(by="uplift_vs_base_%", ascending=False)

    print("\n📊 Top Revenue Shifting Events Across All Revenue Centers:")
    print(final_summary[["RevenueCenter", "Feature", final_summary.columns[0], "mean", "uplift_vs_base_%"]].head(30))

    final_summary.to_csv(os.path.join(output_dir, "uplift_summary_all_centers.csv"), index=False)
    print("\n📁 Global summary saved to 'uplift_summary_all_centers.csv'")
else:
    print("❌ No valid uplift data found.")
```

```python
# ------------------------------------------------------------
# 4. PRINT PER-CENTER UPLIFT TABLES (TOP 5 EVENTS EACH)
# ------------------------------------------------------------

for file in os.listdir(output_dir):
    if file.startswith("uplift_") and file.endswith(".csv") and "all_centers" not in file:
        print("\n" + "="*70)
        print(f"📌 Top Uplift Events for {file.replace('uplift_', '').replace('.csv', '')}")
        print("="*70)

        center_df = pd.read_csv(os.path.join(output_dir, file))
        top_uplifts = (
            center_df.sort_values("uplift_vs_base_%", ascending=False)
            .groupby("Feature", as_index=False)
            .first()
            .sort_values("uplift_vs_base_%", ascending=False)
        )
        print(top_uplifts[["Feature", "mean", "uplift_vs_base_%"]].to_string(index=False))
```
